Client's goal: From the given spreadsheet for every possible match between combination of Code1 and Code2 columns
(must be next to each other, like A1, B2, C3 ...) and columns Unnamed 5 - Unnamed 13 add sum of number character from each combination
match to Count column. 
Ex: in Line 0, columns Code1+Code2 = 'A1' and columns Unnamed5+Unnamed9 = 'A1' which forms a match. For each of combination
matches per line sum all number characters and insetr it in Count column for respective line

```python
import pandas as pd

# importing file
file = pd.read_excel('Upwork Request.xlsx')
file.head(6)
```

```python
# seperating Code1 and Code 2 columns into its own dataframe as Combination
file1 = file.iloc[:,0:2]
file1.head()
```

```python
# seperating options into seperate detaframe and coverting number's dtypes from str to int

file2 = file.iloc[0:5,5:14]
file2.iloc[:,4:] = file2.iloc[:,4:].astype(int)
file2
```

```python
# Joining Code1 and Code2 together as 1 element within the list
file_list = []
for a, i in file1.iterrows():
    file_list.append(f"{i['Code1']}{i['Code2']}")

# Itirating through each line of options dataframe
final_list = []
for row in file2.iterrows():
    p = 0
    g = 0
    combo_list = []
    this_list = []
    temp_list = []
    
    #Seperating options dataframe to letters and numbers lists
    for element in row[1]:
        temp_list.append(element)
    one_list = temp_list[:4]
    two_list = temp_list[4:]
    
    #Creating list of all possible combinations
    for f in range(0,4):
        for n in range(0,5):
            combo_list.append(f'{one_list[g]}{two_list[p]}') 
            p+=1
        p = 0
        g+=1
        
    # Selecting all matches between source Combination and Target list of all Combinations
    # For every match, add its number character to a new list (this_list)
    for t in range(len(file_list)):
        if file_list[t] in combo_list:
            this_list.append(file_list[t][1])
            t+=1
    
    # converting list elements to intigers and adding them into one number and adding that number into final_list
    v = 0
    for i in this_list:
        v+=int(i)        
    final_list.append(v) 
    
    #Cycle repeats for every line in options dataframe (file2)

#result set
final_list
```

```python
# joining result to main dataframe (options dataframe)
file2['result'] = final_list

# vuala..
file2
```
